```python
import polars as pl 
import matplotlib.pyplot as plt 
import os

# Check if the environment variable in .env was actually set in the shell
print(f"PYTHONPATH: {os.environ.get('PYTHONPATH')}")

pl.Config.set_tbl_rows(-1) 
```

```python
from read_data import read_dataframe_from_bed

DF_METH_IMPRINTED_CANDIDATES = read_dataframe_from_bed(
    '/scratch/ucgd/lustre-labs/quinlan/data-shared/tapestry-tools/CEPH1463.GRCh38.hifi.founder-phased/imprinted-candidates.all-samples.meth.bed'
)
DF_METH_IMPRINTED_CANDIDATES
```

```python
from get_samples_and_paths import get_samples_and_paths

def plot_mat_vs_pat_meth(df, sample_meth_beds, mode): 
    df = df.sample(8, seed=42)
    sample_ids, _ = get_samples_and_paths(sample_meth_beds, testing=False)

    for locus in df.iter_rows(named=True):
        chrom, start, end = locus['chrom'], locus['start'], locus['end']
        coordinate = f"{chrom}:{start}-{end}"
        meths_pat, meths_mat, haps_pat, haps_mat, sample_ids_with_meth = [], [], [], [], []
        for sample_id in sample_ids:
            try:
                meths_pat.append(locus[f'{sample_id}_{mode}_based_meth_pat'])
                meths_mat.append(locus[f'{sample_id}_{mode}_based_meth_mat'])
                haps_pat.append(locus[f'{sample_id}_founder_pat'])
                haps_mat.append(locus[f'{sample_id}_founder_mat'])
                sample_ids_with_meth.append(sample_id)
            except KeyError: 
                continue

        for meth_pat, meth_mat, hap_pat, hap_mat, sample_id in zip(meths_pat, meths_mat, haps_pat, haps_mat, sample_ids_with_meth): 
            if meth_pat is not None and meth_mat is not None: 
                print(f"{meth_pat:.2f}", f"{meth_mat:.2f}", hap_pat, hap_mat, sample_id, coordinate)

        plt.figure(figsize=(5,5))
        plt.scatter(
            meths_pat, 
            meths_mat,
            s=150,          
            marker='o',     
            edgecolor='black', 
            linewidths=2,   
            facecolor='red', 
            alpha=0.6   
        )
        plt.title(coordinate)
        plt.xlim(-0.05, 1.05)
        plt.ylim(-0.05, 1.05)
        plt.axvline(x=0, color='k', linestyle='-', linewidth=1, alpha=0.2)
        plt.axhline(y=0, color='k', linestyle='-', linewidth=1, alpha=0.2)
        plt.xlabel(f'paternal {mode}-based methylation')
        plt.ylabel(f'maternal {mode}-based methylation')
        plt.show()
        
plot_mat_vs_pat_meth(
    DF_METH_IMPRINTED_CANDIDATES, 
    sample_meth_beds='/scratch/ucgd/lustre-labs/quinlan/data-shared/tapestry-tools/CEPH1463.GRCh38.hifi.founder-phased/all-cpgs.all-samples.tsv', 
    mode='count'
)
```
